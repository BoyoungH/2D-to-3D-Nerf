```python
def replace_in_file(file_path, old_str, new_str):
    # 파일 읽어들이기
    fr = open(file_path, 'r')
    lines = fr.readlines()
    fr.close()

    # old_str -> new_str 치환
    fw = open(file_path, 'w')
    for line in lines:
        fw.write(line.replace(old_str, new_str))
    fw.close()
```

```python
import numpy as np
```

```python
num = len(camera_data_split)
print(num)
```

```python
final_images_data = images_df.loc[:,['QW','QX','QY','QZ','TX','TY','TZ']]
numpy_images_data = images_df[['QW','QX','QY','QZ','TX','TY','TZ']].values.tolist()
numpy_images_data[32]
```

```python
# 카메라 pose 값을 나타내는 변수
qw=[]
qx=[]
qy=[]
qz=[]
tx=[]
ty=[]
tz=[]
for i in range(num):
    qw.append(numpy_images_data[i][0])
    qx.append(numpy_images_data[i][1])
    qy.append(numpy_images_data[i][2])
    qz.append(numpy_images_data[i][3])
    tx.append(numpy_images_data[i][4])
    ty.append(numpy_images_data[i][5])
    tz.append(numpy_images_data[i][6])
```

```python
tx = [float (i) for i in tx]
ty = [float (i) for i in ty]
tz = [float (i) for i in tz]
```

```python
qw=list(map(float, qw))
qx=list(map(float, qx))
qy=list(map(float, qy))
qz=list(map(float, qz))
tx=list(map(float, tx))
ty=list(map(float, ty))
tz=list(map(float, tz))

print(type(qw[0]))
```

```python
# 회전 쿼터니언으로부터 회전 행렬을 구함
R = []
for i in range(num):
  R.append(np.array([[1-2*qy[i]**2-2*qz[i]**2, 2*qx[i]*qy[i]-2*qz[i]*qw[i], 2*qx[i]*qz[i]+2*qy[i]*qw[i]],
                [2*qx[i]*qy[i]+2*qz[i]*qw[i], 1-2*qx[i]**2-2*qz[i]**2, 2*qy[i]*qz[i]-2*qx[i]*qw[i]],
                [2*qx[i]*qz[i]-2*qy[i]*qw[i], 2*qy[i]*qz[i]+2*qx[i]*qw[i], 1-2*qx[i]**2-2*qy[i]**2]]))
```

```python
# 변환 행렬을 구함
T = []
P = []
for i in range(num):
  T.append(np.array([tx[i], ty[i], tz[i]]).reshape(-1, 1))
for i in range(num):
  P.append(np.hstack([R[i], T[i]]))
for i in range(num):
  P[i] = np.vstack([P[i], [0.0, 0.0, 0.0, 1.0]])
```

```python
import cv2
import os

imgrgb=[]
image_array=[]
# 이미지가 저장된 경로
image_dir = '/content/scan1'

# 이미지 파일 리스트 가져오기
image_files = os.listdir(image_dir)

# 이미지 파일마다 RGB 값 추출
for file in image_files:
    # 이미지 로드
    image_path = os.path.join(image_dir, file)
    image = cv2.imread(image_path)

    # 이미지를 numpy 배열로 변환
    image_array[file] = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    for i in range(image_array[file].shape[0]):
      for j in range(image_array[file].shape[1]):
        # (i, j) 위치의 픽셀 값 (RGB)
          imgrgb[file] = image_array[file][i, j]

print(imgrgb[1])
```

```python
from PIL import Image
import os

# 이미지 파일들이 들어 있는 폴더 경로
folder_path = '/content/scan1'

# 모든 이미지에서 추출한 RGB 값 저장할 리스트
imgrgb = []

# 폴더 내 모든 이미지 파일에 대해 반복문 실행
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # 이미지 파일 열기
        image_path = os.path.join(folder_path, filename)
        image = Image.open(image_path)

        # 이미지 RGB 값 추출
        rgb_values = list(image.getdata())

        # 추출한 RGB 값을 imgrgb 배열에 추가
        imgrgb.extend(rgb_values)

print(f"Total RGB values: {len(imgrgb)}")
print(imgrgb[0])
```
