```python
import cv2
vidcap = cv2.VideoCapture('/content/IMG_8687.MOV')
success,image = vidcap.read()
count = 0
while success:
  cv2.imwrite("/content/drive/MyDrive/images/%05d.png" % count, image)     # save frame as PNG file
  success,image = vidcap.read()
  print('Read a new frame: ', success)
  count += 1

print("finish! convert video to frame")
```

```python
import cv2

vidcap = cv2.VideoCapture('/content/IMG_8687.MOV')
frame_rate = 8  # 초당 프레임 수
success, image = vidcap.read()
count = 0
frame_count = 0

while success:
    if frame_count % frame_rate == 0:
        cv2.imwrite("/content/drive/MyDrive/imagesssss/%05d.png" % count, image)     # save frame as PNG file
        count += 1

    success, image = vidcap.read()
    frame_count += 1

    print('Read a new frame: ', success)

print("Finish! Video frames converted")
```

```python
import os
import glob
from PIL import Image

files = glob.glob('/content/drive/MyDrive/imagesssss/*.png')

for f in files:
    img = Image.open(f)
    img_resize = img.resize((int(200), int(200)))
    title, ext = os.path.splitext(f)
    img_resize.save(title + ext)
print("done!")
```
